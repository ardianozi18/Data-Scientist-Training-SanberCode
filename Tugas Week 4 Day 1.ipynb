Apa itu machine learning?
Machine Learning adalah kemampuan komputer untuk mempelajari suatu keadaan untuk dilakukan atau bahkan dilakukan perbaikan di kemudian waktu.

Apa itu data feature dan data target?
Data Feature atau Predictor merupakan data variable untuk melakukan prediksi.
Sedangkan Data Target atau Label merupakan hasil dari prediksi tersebut.


Apa Perbedaan Supervised Learning, dan Unsupervised Learning?
Supervised Learning menggunakan data latih (data yang sudah berlabel) untuk pembelajaran. Tujuannya untuk prediksi dan klasifikasi.
Unsupervised Learning tidak menggunakan data latih dan tidak memiliki variabel target. Tujuannya untuk clustering data.


Apa Jenis2 yang ada di dalam Supervised Learning? Jelaskan Perbedaannya!
Ada Classification dan Regression.
Classification outputnya merupakan labelling, misal nya cuaca hari ini apakah "mendung", "berawan", atau "hujan"
Regression merupakan nilai yang linear atau continues hasilnya.

Apa perbedaan Hyperparameter dan Parameter?
Hyperparameter didefinisikan di awal dalam pembuatan objek model. Tidak ada sangkut pautnya dengan model. Untuk mengoptimisasi model.
Parameter dihasilkan di akhir. Ada sangkut pautnya dengan model. Ada di dalam model.


Sebutkan step2 dalam mengaplikasikan algoritma apapun dalam machine learning?
1. ilih model atau algoritma yang akan digunakan
2. Pilih Hyperparameter ketika membuat objek model
3. Pisahkan antara Data Feature dan Data Target - > data diinput dalam matriks bukan vektor
4. Perintahkan model untuk mempelajari data menggunakan method .fit()
5. Aplikasikan model ke dalam test data dengan menggunakan method .transform() dan .predict() untuk unsupervized learning



<h1>Quiz 2 : Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian applikasikan untuk membuat model bagi data (variable x, y) di bawah ini. kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
#1 Pilih Model atau Algorithm yang aka digunakan dalam pattern
from sklearn.linear_model import LinearRegression
```

```python
#2 Pilih Hyperparameter ketika membuat objek model
model_lr = LinearRegression(fit_intercept=True)
```

```python
#3 Pisahkan antara Data Feature dan Data Target, ingat input dalam matrix bukan vektor
x_matriks = x[:, np.newaxis]
```

```python
#4 perintahkan model untuk mempelajari data menggunakan method .fit()
model_lr.fit(x_matriks, y)
```

```python
#5 Aplikasikan model ke dalam test data menggunakan method .predict() untuk supervised learning atau method .transform() dan .predict() untuk unsupervized learning
x_test = np.linspace(20, 30, 5)
```

```python
x_test_matriks = x_test[:, np.newaxis]
```

```python
y_test = model_lr.predict(x_test_matriks)
```

```python
#Data sebenarnya, bukan test lagi
y_train = model_lr.predict(x_matriks)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y, color='blue', label='Data Asli')
ax.scatter(x, y_train, color='orange', label='Prediksi terhadap Data Training')
ax.plot(x_test_matriks, y_test, color='green', label='Prediksi terhadap Data Baru')
plt.legend(bbox_to_anchor=(0.7, 0.8, 0.5, 0.5))
plt.show
```

Contoh Jawaban Hasil Pembuatan Model dengan Algoritma DecisionTreeRegressor (Jawaban tidak haru sama) :



```python
#1 Pilih Model atau Algorithm yang aka digunakan dalam pattern
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
#2 Pilih Hyperparameter ketika membuat objek model
model_dt = tree.DecisionTreeRegressor(criterion='mse', max_depth=2)
#3 Pisahkan antara Data Feature dan Data Target, ingat input dalam matrix bukan vektor
x_matriks = x[:, np.newaxis]
#4 perintahkan model untuk mempelajari data menggunakan method .fit()
model_dt.fit(x_matriks, y)
#5 Aplikasikan model ke dalam test data menggunakan method .predict() untuk supervised learning atau method .transform() dan .predict() untuk unsupervized learning
x_test = np.linspace(20, 30, 15)
x_test_matriks = x_test[:, np.newaxis]
y_test = model_dt.predict(x_test_matriks)
#Data sebenarnya, bukan test lagi
y_train = model_dt.predict(x_matriks)
tree.plot_tree(model_dt.fit(x_matriks, y))
```
