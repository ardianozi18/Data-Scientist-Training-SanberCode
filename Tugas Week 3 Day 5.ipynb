```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Quiz 1 Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
mean_score = df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean()
mean_score
```

<h2>Perbandingan Kuantitatif Barplot : Grouping Visualisasi Dengan Barplot</h2>

```python
#data
x = np.arange(4)
y = mean_score.values

#membuat figure
fig, ax = plt.subplots(figsize=(20,10))
lebar = 0.07
genre = list(mean_score.index)

for i in range(len(genre)):
  ax.bar(x+(i-1)*lebar, y[i], width = lebar, label = genre[i])

#membuat nilai sumbu x
ax.set_xticklabels([0] + list(mean_score.columns))
#membuat label sumbu x
ax.set_xlabel('Region Sales')
#membuat label sumbu y
ax.set_ylabel('Mean Sales')
#judul grafik
ax.set_title('Mean Sales Video Games by Genre')
#agar label sumbu x sesuai
plt.xticks(x + 0.5, mean_score.columns)
#background
plt.style.use('ggplot')

plt.legend(bbox_to_anchor=(0.56, 0.64, 0.5, 0.5))
plt.show()
```

Insight : Rata-rata penjualan tertinggi adalah games ber-genre Platform untuk region North America. Untuk Region Eropa rata-rata penjualan tertinggi untuk games genre Shooter. Untuk Region Jepang rata-rata penjualan tertinggi adalah Role-Playing.

<h2>Perbandingan Kuantitatif Barplot : Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#data
x = np.arange(4)
y = mean_score.values

#membuat figure
fig, ax = plt.subplots(figsize=(20,10))
lebar = 0.8
genre = list(mean_score.index)
bottom = [0,0,0,0]

for row in mean_score.iterrows():
  ax.bar(x, row[1], bottom = bottom, label = row[0])
  bottom = bottom + row[1]

#membuat nilai sumbu x
ax.set_xticklabels([0] + list(mean_score.columns))
#membuat label sumbu x
ax.set_xlabel('Region Sales')
#membuat label sumbu y
ax.set_ylabel('Mean Sales')
#judul grafik
ax.set_title('Mean Sales Video Games by Genre')
#agar label sumbu x sesuai
plt.xticks(x + 0.03, mean_score.columns)
#background
plt.style.use('ggplot')

plt.legend(bbox_to_anchor=(0.56, 0.64, 0.5, 0.5))
plt.show()
```

Insight : Rata-rata penjualan tertinggi adalah games ber-genre Platform untuk region North America. Untuk Region Eropa rata-rata penjualan tertinggi untuk games genre Shooter. Untuk Region Jepang rata-rata penjualan tertinggi adalah Role-Playing.
